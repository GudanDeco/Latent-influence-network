```python
from datetime import datetime, timedelta

def get_tweet_real_time(snowflake_id):
    """
    Get millisecond timestamp from snowflake ID.
    """
    snowflake_id = int(snowflake_id)
    timestamp_ms = (snowflake_id >> 22) + 1288834974657  
    return datetime.utcfromtimestamp(timestamp_ms / 1000.0)
```

```python
###################################################
import networkx as nx
import random
from gensim.models import Word2Vec
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def get_deep_walk_result(graph):

    class DeepWalk:
        def __init__(self, graph, walk_length, num_walks, workers=1):
            self.graph = graph
            self.walk_length = walk_length
            self.num_walks = num_walks
            self.workers = workers

        def random_walk(self, start_node):
            walk = [start_node]
            for _ in range(self.walk_length - 1):
                cur = walk[-1]
                neighbors = list(self.graph.neighbors(cur))
                if len(neighbors) > 0:
                    walk.append(random.choice(neighbors))
                else:
                    break
            return [str(node) for node in walk]

        def generate_walks(self):
            walks = []
            nodes = list(self.graph.nodes())
            for _ in range(self.num_walks):
                random.shuffle(nodes)
                for node in nodes:
                    walks.append(self.random_walk(node))
            return walks

        def train(self, embed_size=128, window_size=5):
            walks = self.generate_walks()
            model = Word2Vec(walks, vector_size=embed_size, window=window_size, workers=self.workers)
            return model

    deepwalk = DeepWalk(graph, walk_length=8, num_walks=160, workers=4)

    model = deepwalk.train(embed_size=256, window_size=10)

    node_ids = list(model.wv.index_to_key)

    # Create the encoder
    encoder = LabelEncoder()
    # Fit the encoder on the labels
    user_labels = pd.read_csv('User_label_16.csv')
    user_labels['topic'] = encoder.fit_transform(user_labels['Label'])

    # Prepare data
    X_train = []
    y_train = []
    X_test = []
    y_test = []
    
    for i, row in user_labels.iterrows():
        user_id_str = str(row['Node'])
        if user_id_str in model.wv:
            if row['Train_test_split'] == 'Train':
                X_train.append(model.wv[user_id_str])
                y_train.append(row['topic'])
            elif row['Train_test_split'] == 'Test':
                X_test.append(model.wv[user_id_str])
                y_test.append(row['topic'])
                
    # Standardize the data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    # Create a logistic regression model
    log_model = make_pipeline(StandardScaler(), LogisticRegression(multi_class='ovr', solver='liblinear'))

    # Train the model
    log_model.fit(X_train, y_train)

    # Create a SVM model with a RBF kernel
    svm_model = make_pipeline(StandardScaler(), SVC(kernel='rbf', gamma='auto',probability=True))

    # Train the model
    svm_model.fit(X_train, y_train)

    logist_stance_predictions = log_model.predict(X_test)
    svm_stance_predictions = svm_model.predict(X_test)
    logist_stance_probabilities = log_model.predict_proba(X_test)
    svm_stance_probabilities = svm_model.predict_proba(X_test)

    logist_precision_stance = precision_score(y_test, logist_stance_predictions, average='weighted', zero_division=1)
    logist_recall_stance = recall_score(y_test, logist_stance_predictions, average='weighted')
    logist_f1_stance = f1_score(y_test, logist_stance_predictions, average='weighted')
    logist_auc_roc_stance = roc_auc_score(y_test, logist_stance_probabilities, multi_class='ovr', average='weighted')

    svm_precision_stance = precision_score(y_test, svm_stance_predictions, average='weighted', zero_division=1)
    svm_recall_stance = recall_score(y_test, svm_stance_predictions, average='weighted')
    svm_f1_stance = f1_score(y_test, svm_stance_predictions, average='weighted')
    svm_auc_roc_stance = roc_auc_score(y_test, svm_stance_probabilities, multi_class='ovr', average='weighted')

    print(f"Finished: Here is logist precision: {logist_precision_stance}")
    
    return {
        "logist_precision_stance": logist_precision_stance,
        "logist_recall_stance": logist_recall_stance,
        "logist_f1_stance": logist_f1_stance,
        "logist_auc_roc_stance": logist_auc_roc_stance,

        "svm_precision_stance": svm_precision_stance,
        "svm_recall_stance": svm_recall_stance,
        "svm_f1_stance": svm_f1_stance,
        "svm_auc_roc_stance": svm_auc_roc_stance,
        "Train_len":len(X_train),
        "Test_len":len(X_test)
    }
```

```python
import math
from tqdm import tqdm
time_list = [5,10,15,20,25,30,35,40,45,50,55,60]
weight_list = [0,1,2]
folder = r'./lcn_tweet_16_weber/'
for time in time_list:
    print("=======================================================================================")
    print("Start Time:",time)
    LIN = nx.Graph()
    time_block_size_seconds = time * 60 

    min_time = LCN_csv['parsed_real_time'].min()
    max_time = LCN_csv['parsed_real_time'].max()
    total_duration_seconds = (max_time - min_time).total_seconds()

    num_blocks = math.ceil(total_duration_seconds / time_block_size_seconds)

    block_ranges = {i: [] for i in range(num_blocks)}

    for index, row in LCN_csv.iterrows():
        if pd.notnull(row['parsed_real_time']):
            time_diff_seconds = (row['parsed_real_time'] - min_time).total_seconds()
            block_index = int(time_diff_seconds / time_block_size_seconds)
            block_ranges[block_index].append(index)

    for block_index, indices in block_ranges.items():

        for i in range(len(indices)):
            for j in range(i + 1, len(indices)):
                index_i = indices[i]
                index_j = indices[j]

                if LCN_csv.loc[index_i, 'LCN_target'] == LCN_csv.loc[index_j, 'LCN_target']:
                    u = LCN_csv.loc[index_i, 'Target']
                    v = LCN_csv.loc[index_j, 'Target']
                    if LIN.has_edge(u, v):
                        LIN[u][v]['weight'] += 1
                    else:
                        LIN.add_edge(u, v, weight=1)

    for node in LIN.nodes():
        LIN.nodes[node]['most_common_label'] = label_data.get(str(node), None) 

    for weight in weight_list:
        print("#####")
        print("    Start weight:",weight)
        edges_to_remove = [(u, v) for u, v, attr in LIN.edges(data=True) if attr.get('weight', 1) <= weight]
        LIN_remove = LIN.copy()
        print("    Total edges:", LIN.number_of_edges())
        print("    Need to remove: ",len(edges_to_remove))
        print("    Remain:",LIN.number_of_edges() - len(edges_to_remove))
        for edge in edges_to_remove:
            LIN_remove.remove_edge(*edge)

        nx.write_graphml(LIN_remove,folder + f'k_core4_LCN_{time}_{weight}.graphml')

        result_list = []
        for i in range(10):
            result_list.append(get_deep_walk_result(LIN_remove))
        with open(r'./lcn_tweet_16_weber/' + f'k_core4_LCN_{time}_{weight}.pkl', 'wb') as file:
            pickle.dump(result_list, file)
        print("Final_result for", time)
        print_acc(result_list)
```

```python
import os
import pickle
import matplotlib.pyplot as plt

def read_and_process_pkl_files(folder_path):
    results = {}  # {time: {weight: {'log_acc': [], 'svm_acc': []}}}
    
    for file in os.listdir(folder_path):
        if file.endswith(".pkl") and file != "DRN.pkl":
            with open(os.path.join(folder_path, file), 'rb') as f:
                parts = file.split('_')
                print(parts)
                time = int(parts[-2])
                weight = int(parts[-1][0])
                if time < 65:
                    result_list = pickle.load(f)
                    log_acc, svm_acc = [], []
                    for result in result_list:
                        log_acc.append(result["logist_precision_stance"])
                        svm_acc.append(result["svm_precision_stance"])

                    log_average = sum(log_acc) / len(log_acc)
                    svm_average = sum(svm_acc) / len(svm_acc)

                    if time not in results:
                        results[time] = {}
                    results[time][weight] = {'log_acc': log_average, 'svm_acc': svm_average}

    return results

```
